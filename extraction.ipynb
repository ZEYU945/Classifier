```python
import numpy as np
import pandas as pd
import os
import cv2
```

```python
df = pd.read_excel('cellule_expert_resultats.xlsx')
df.head()
```

```python
Index = df['ID'].values.tolist()
index = []
for n in Index:
    component = str(n)+'.jpg'
    index.append(component)
index
```

```python
PROJECT_ROOT = os.path.join(os.path.dirname(os.path.abspath('')),'extraction','files')
Images = []
for filename in os.listdir(PROJECT_ROOT):
    if filename in index:
        img = cv2.imread(os.path.join(PROJECT_ROOT,filename))
        save = os.path.join(os.path.dirname(os.path.abspath('')),'extraction','files_choose',filename)
        cv2.imwrite(save, img)
```
